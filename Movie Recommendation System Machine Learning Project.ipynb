```
############Splitting Data into Training and Test for Training

(training,test) = ratings.randomSplit([0.8, 0.2], seed=100)

training = training.unionAll(myRatings)
```

```
# PART 4 - TRAINING THE MODEL 
```

```
# 7)  Using ALS  machine learning algorithm model to train in MLib library to train the dataset

from pyspark.ml.recommendation import ALS
als = ALS(maxIter=5, regParam=0.15, userCol="userId", itemCol="movieId", ratingCol="rating", seed=100)
model = als.fit(training)
```

```
######################################## PART 5 - EVALUATING THE MODEL #################################
```

```
# Using the dropna coomand to remove those user that did not make rate the movies 

predictions = model.transform(test).dropna()        # prediction is generated by using transform 
predictions .show()
```

```
# Using Root Mean Square Error as evaluating metric 

from pyspark.ml.evaluation import RegressionEvaluator

evaluator = RegressionEvaluator(metricName="rmse", labelCol="rating", predictionCol="prediction")
rmse = evaluator.evaluate(predictions)
print('Root Mean Square Error is %g' %rmse)
```

```
from pyspark.ml.tuning import ParamGridBuilder
# Creating a hyperparamter grid
parameters = ParamGridBuilder()\
.addGrid(als.maxIter,[5,5,5,5,5,5,5])\
.addGrid(als.regParam,[0.15,0.16,0.18,0.2,0.1,0.25,0.01])\
.build()
```

```
from pyspark.ml.tuning import TrainValidationSplit
tvs = TrainValidationSplit()\
.setSeed(100)\
.setTrainRatio(0.8)\
.setEstimatorParamMaps(parameters)\
.setEstimator(als)\
.setEvaluator(evaluator)
```

```
# Training model using grid search 
gridsearchModel = tvs.fit(training)
```

```
# Selecting the best model and identify parameters
bestModel = gridsearchModel.bestModel
print("Parameter for the best model:")

print("MaxIter Parameter : %g" %bestModel._java_obj.parent().getMaxIter())
print("RegParam Parameter : %g" %bestModel._java_obj.parent().getRegParam())
```

```
# Using the best model to make prediction on the test dataset

evaluator.evaluate(bestModel.transform(test))
```

```
mySampledMovies = model.transform(myRatings)
myRmse = evaluator.evaluate(mySampledMovies)
print("Root Mean Square Error is %g" %myRmse)
mySampledMovies.show()
```

```
###################################### PART 6 - USING THE MODEL TO GENERATE MOVIE RECOMMENDATIONS  #######################################
```

```
# 10) We have to create a new DataFrame which contains movvie ids and columns popluated with user id 0s 

from pyspark.sql import functions
myGeneratedPredictions = movies.withColumn("userId", functions.expr("int('0')"))
myGeneratedPredictions = model.transform(myGeneratedPredictions)
myGeneratedPredictions = myGeneratedPredictions.dropna()
myGeneratedPredictions.orderBy("prediction", ascending=False).show(truncate=False)
```

```
# USING THE OUR MODEL TO GENERATES RECOMMENDATIONS FOR ALL THE USERS 

userRecs = model.recommendForAllUsers(10)
```

```
userRecs.show(truncate=False)
```

```
# Returning recommendation for a specific user

from pyspark.sql.functions import explode

userRecs.where(userRecs.userId==1).select("recommendations") \
    .withColumn("recommendations", explode("recommendations"))\
    .select("recommendations.movieId", "recommendations.rating")\
    .join(movies, ["movieId"]) \
    .show(truncate=False)
```
